```python
def Contact_Book():
    flag = 4
    while flag:
        print('''
        .....Welcome to Contact Book Mangement.....
        ...........To Open Press Y/N..............
        ''')
        open1 = input('Y/N: ').title()
        if open1 == 'Y':
            print('''
            To Open Contact Book                 Press 1
            To Search Contact                    Press 2
            To Add New Contact                   Press 3
            To Update Name/Number/Email/NickName Press 4
            To Delete Contact                    Press 5
            To Exit                              Press 0
            ''')
            Option1 = int(input('Option: '))
            if Option1 == 1:
                print(contact_data)

            elif Option1 == 2:
                search_data()

            elif Option1 == 3:
                add_details()

            elif Option1 == 4:
                update_data()

            elif Option1 == 5:
                delete_data()
            
            elif Option1 == 0 :
                break
                
            else:
                print(f'Please Try Again You Got {flag - 1} chance')
                flag-=1

        else:
            print('Thank You!!!!!!')
            flag-=1
        
```
